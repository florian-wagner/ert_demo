This interactive [Jupyter notebook](www.jupyter.org) uses [pyGIMLi](www.pygimli.org) to interactively evaluate the imaging capabilities of an electrical resistivity survey with respect to a circular anomaly in the subsurface. We start by importing the necessary [Python](www.python.org) packages.

**Hint**: Press `Run` in the toolbar at the top or hit `Shift-Enter` on your keyboard to evaluate each cell.

*Note*: For this tutorial, knowledge of the Python programming language is not necessary. But if you are interested, the [Scipy Lecture Notes](https://www.scipy-lectures.org/) provide a very good starting point to learn scientific programming with Python.

```python
%matplotlib inline
from ert import *
```

## Change the position, depth, and radius of the circular anomaly by moving the sliders.

```python
geometry = interact(create_geometry, position=(-30,30,5), depth=(6,15,1), radius=(2,5,1), continous_update=False)
```

## The subsurface model needs to be discretized for numerical forward simulation.

```python
global mesh, scheme
mesh, scheme = create_mesh(geometry.widget.result)
```

## Change the electrical resistivity $\rho$ (in $\Omega$m) of the background and the anomaly by moving the sliders and see how this affects the simulated data in the pseudosection.

```python
dataset = interact(simulate, res_background=(10,400,50), res_anomaly=(10,500,50), mesh=fixed(mesh), scheme=fixed(scheme))
```

## Finally, we can run an inversion on the dataset and see how well the electrical resistivity method is able to reconstruct the circular anomaly. Go back to the top and change the geometry to get a feeling for the sensitivity and resolution limits.


```python
invert(dataset.widget.result)
```
